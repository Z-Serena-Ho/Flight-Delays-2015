```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

df_csv = pd.read_csv(r'C:/Users/zsere/Documents/Raw Data/Flight Delays 2015/flights.csv')

#Truncated dataframe.
df_trim = df_csv[['MONTH', 'DAY', 'DAY_OF_WEEK', 'AIRLINE', 'ORIGIN_AIRPORT', 'DESTINATION_AIRPORT']]

total_flights = len(df_trim.index)
```

```python
df_month = df_trim.groupby('MONTH').size()
sns.barplot(x = df_month.index, y = df_month.values)
```

```python
df_airlines = df_trim.groupby('AIRLINE')['AIRLINE'].count().reset_index(name='DELAYS')
df_airlines.sort_values('DELAYS', inplace = True)

df_code_chart = pd.read_csv(r'C:/Users/zsere/Documents/Raw Data/Flight Delays 2015/airlines.csv')
df_airlines.rename(columns = {'AIRLINE': 'IATA_CODE'}, inplace = True)
df_airlines = pd.merge(df_airlines, df_code_chart)

print(df_airlines[['IATA_CODE', 'AIRLINE']])
sns.barplot(df_airlines, x = 'IATA_CODE', y = 'DELAYS')
```
